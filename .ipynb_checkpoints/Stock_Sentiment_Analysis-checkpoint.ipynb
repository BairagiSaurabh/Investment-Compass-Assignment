# Use News api to get all news related to a stock for a specified range of date

```python
from newsapi import NewsApiClient
my_api_key = ""
newsapi = NewsApiClient(api_key=my_api_key)
```

```python
#newsapi.get_sources()
```

```python
def get_news_articles(query,page):

    all_articles = newsapi.get_everything(q=query,
                                      sources='bbc-news,the-verge,abc-news',
                                      domains='bbc.co.uk,techcrunch.com,abcnews.go.com',
                                      from_param='2024-02-20',
                                      to='2024-03-12',
                                      language='en',
                                      sort_by='relevancy',
                                      page=page)
    title = []
    description = []
    date = []
    for article in (all_articles['articles']):
        title.append(article['title'].strip())
        description.append(article['description'].strip())
        date.append(article['publishedAt'].strip())
    df = pd.DataFrame({'Date' : date, 'Description' : description,'Title':title})
    return df
```

```python
def get_df(company):
    company = company
    dfs = []
    for i in tqdm(range(1,5)):
        d1 = get_news_articles(company,i)
        dfs.append(d1)
        if len(d1) < 2:
            break
    main_df = pd.concat(dfs, ignore_index=True)
    main_df['company'] = company
    
    return main_df
```

```python
main_df_amazon = get_df("Amazon")
main_df_apple = get_df("Apple")
main_df_sp500 = get_df("S&P 500")
```

```python
all_df = pd.concat([main_df_amazon,main_df_apple,main_df_sp500],axis=0) # we combine all dataframes into one
```

# Data Cleaning

```python
import nltk
nltk.download('stopwords')
```

```python
all_df['text'] = all_df['Title'] + ' ' + all_df['Description'] # combining both textual field into one
```

```python
lemma = WordNetLemmatizer() # applying lemmatization to return dictionary form of words

all_stopwords = stopwords.words('english') # this consists all the stopwords, which will be removed later.

# Removing the following words from list containing stopwords
all_stopwords.remove('not')
all_stopwords.remove('against')

def clean_text(text):
    """
    This function removes punctuations, stopwords and other non alpha numeric characters.
    We expand the contractions and replace some words by an empty string
    
    """
    statement = text.lower().strip()
    statement = statement.replace("won't", "will not").replace("cannot", "can not").replace("can't", "can not")\
                           .replace("n't", " not").replace("what's", "what is").replace("it's", "it is")\
                           .replace("'ve", " have").replace("i'm", "i am").replace("'re", " are")\
                           .replace("he's", "he is").replace("she's", "she is").replace("*****"," ")\
                           .replace("%", " percent ").replace("₹", " rupee ").replace("$", " dollar ")\
                           .replace("€", " euro ").replace("'ll", " will").replace("doesn't", "does not")
    
    statement = re.sub('[^a-zA-Z]',' ',statement) # replacing whatever isn't letters by an empty string
    statement = statement.split() # forming list of words in a given review
    final_statement = [lemma.lemmatize(word) for word in statement if not word in set(all_stopwords)]
    final_statement_ = ' '.join(final_statement) # joining the words and forming the review again without stopwords
    return final_statement_  
```

```python
all_df['clean_text'] = all_df['text'].apply(lambda x : clean_text(x)) 
```

```python
all_df.drop(['text'],axis=1,inplace=True)
```

# Derive Sentiments from Text data

```python
def sentiment_scores(sentence):
    senti = SentimentIntensityAnalyzer() 
    sentiment_dict = senti.polarity_scores(sentence) # this line returns the polarity for sentence
  
    if sentiment_dict['compound'] >= 0.05 : # if the compound score is >= 0.05 then the review is positive
        return ("Positive"),sentiment_dict['pos'],sentiment_dict['compound']

    elif sentiment_dict['compound'] <= - 0.05 : # if the compound score is <= -0.05 then the review is negative
        return ("Negative"),sentiment_dict['neg'],sentiment_dict['compound']

    else :
        return ("Neutral"),sentiment_dict['neu'],sentiment_dict['compound'] # if compound score is in between 0.05 and -0.05 then the review is neutral
    
```

```python
def form_final_data(data):
    starttime = timeit.default_timer()
    
    sentiment_ = [] # this list carries the sentiment for the review
    compound = [] # this consists of score based on which the sentiments are defined

    for text in data["clean_text"].values: # applying the above defined function by iterating through each review
        a,b,c = sentiment_scores(text)
        sentiment_.append(a)
        compound.append(c)
        
    data["Sentiment"] = sentiment_ # adding the sentiment to dataframe
    data["Score"] = compound # adding the score to dataframe
    
    data["Date"] = pd.to_datetime(data['Date']) 
    data.sort_values(by = 'Date', inplace = True) # sorting the date in ascending order
    final_dff = data.reset_index().drop(["index"], axis=1) # resetting the index
    print("The time difference is :", timeit.default_timer() - starttime)
    return final_dff
```

```python
final_df = form_final_data(all_df)
```

```python
def plot_company_frequency(df,feature):
    company_counts = df[feature].value_counts()
    plt.figure(figsize=(8, 4))
    company_counts.plot(kind='bar', color='skyblue')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.title(f'Frequency of Each {feature}')
    plt.show()
```

```python
plot_company_frequency(final_df,'company')
```

```python
def plot_sentiments_timeseries(df):
    df['Date'] = pd.to_datetime(df['Date'])
    fig, ax = plt.subplots(figsize=(10, 6))
    for company, data in df.groupby('company'):
        sentiment_counts = data.groupby(['Date', 'Sentiment']).size().unstack(fill_value=0)
        total_counts = sentiment_counts.sum(axis=1)
        sentiment_percentages = sentiment_counts.div(total_counts, axis=0) * 100
        sentiment_percentages.plot(ax=ax, y=['Positive', 'Negative', 'Neutral'], label=company)
    plt.xlabel('Date')
    plt.ylabel('Percentage of Sentiments')
    plt.title('Time Series of Sentiments by Company')
    plt.legend(title='Company')
    plt.show()
```

```python
def lineplot_score(data,feature):
    d1 = data.groupby(by='company')
    d2 = d1.get_group(feature)
    plt.figure(figsize=(14, 4))
    plt.xlabel("Date")
    plt.ylabel('Score')
    plt.title(f'{feature} sentiment over time')
    sns.lineplot(x="Date", y="Score", data=d2)
```

```python
def plot_top_n_words(data,feature,n):
    d1 = data.groupby(by='company')
    d2 = d1.get_group(feature)
    text_data = [i for i in d2.clean_text.values]
    all_text = ' '.join(text_data)
    words = all_text.split()
    word_freq = pd.Series(words).value_counts()
    top_n_words = word_freq.head(n)
    top_n_df = top_n_words.reset_index()
    top_n_df.columns = ['Word', 'Frequency']
    plt.figure(figsize=(14, 6))
    sns.barplot(x='Word', y='Frequency', data=top_n_df, palette='viridis')

    plt.xticks(rotation=45)
    plt.xlabel('Word')
    plt.ylabel('Frequency')
    plt.title(f'Top {n} Words for {feature}')
    plt.show()
```

```python
def wc(data,company,sentiment):
    s1 = data.groupby(by = 'company')
    s2 = s1.get_group(company)
    s3 = s2.groupby(by = 'Sentiment')
    s4 = s3.get_group(sentiment)
    all_words = s4["clean_text"].values
    
    wordcloud = WordCloud(background_color = 'black', width = 512,height = 384,contour_color='black',contour_width=1,
                         colormap='RdYlGn',collocations=True).generate(' '.join(all_words))
    plt.title(f"Wordcloud for {company}'s {sentiment} Sentiment")
    plt.imshow(wordcloud) # image show
    plt.axis('off') # to off the axis of x and y
```

```python
plt.figure(figsize=(11, 5))
plt.subplot(1, 2, 1)
wc(final_df, 'Amazon', "Positive")

plt.subplot(1, 2, 2)
wc(final_df, 'Amazon', "Negative")
plt.tight_layout()
plt.show()
```
