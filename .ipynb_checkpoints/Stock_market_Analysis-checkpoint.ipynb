```python
#import yfinance as yf
import pandas as pd
import os
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import math
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# sp500 = yf.Ticker("^GSPC")
# sp500 = sp500.history(period="max")
# sp500.to_csv("sp500.csv")

sp500 = pd.read_csv("sp500.csv",index_col='Date',parse_dates=True)
sp500.index.freq='MS'
```

```python
sp500
```

```python
sp500 = sp500.iloc[20000:,:]
sp500
```

```python
sp500.isnull().sum()
```

```python
# We will build a model to predict the closing prices.

data = sp500[["Close"]]
```

# Distribution of Closing prices over the years:

```python
data.plot(figsize=(12,6))
```

```python
# Split into train and test set

training_size=int(len(data)*0.85)
test_size=len(data)-training_size

training_size,test_size
```

```python
train_data,test_data=data.iloc[0:training_size,:],data.iloc[training_size:len(data),:1]
```

```python
train_data.shape,test_data.shape
```

```python
train_data.head(5)
```

```python
test_data.tail(5)
```

```python
# Transform the data using normalization

scaler=MinMaxScaler(feature_range=(0,1))
scaler.fit(train_data)
scaled_train = scaler.transform(train_data)
scaled_test = scaler.transform(test_data)
```

```python
scaled_train[:5]
```

```python
scaled_test[:5]
```

```python
# Form proper dataset for LSTM 

def create_dataset(dataset, time_step=1):
	dataX, dataY = [], []
	for i in range(len(dataset)-time_step-1):
		a = dataset[i:(i+time_step), 0]   ###i=0, 0,1,2,3-----99   100 
		dataX.append(a)
		dataY.append(dataset[i + time_step, 0])
	return np.array(dataX), np.array(dataY)
```

```python
# reshape into X=t,t+1,t+2,t+3 and Y=t+4
time_step = 100
X_train, y_train = create_dataset(scaled_train, time_step)
X_test, ytest = create_dataset(scaled_test, time_step)
```

```python
(X_train.shape), (y_train.shape)
```

```python
(X_test.shape), (ytest.shape)
```

```python
# reshape input to be [samples, time steps, features] which is required for LSTM

X_train =X_train.reshape(X_train.shape[0],X_train.shape[1] , 1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1] , 1)
```

```python
# Building model

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(100,1)))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mse',optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,validation_data=(X_test,ytest),epochs=5,batch_size=64,verbose=1)
```

```python
loss_per_epoch = model.history.history['loss']
plt.plot(range(len(loss_per_epoch)),loss_per_epoch)
```

```python
# Predicting on train set

train_predict=model.predict(X_train)
train_predict_norm=scaler.inverse_transform(train_predict)
math.sqrt(mean_squared_error(y_train,train_predict_norm))
```

```python
test_predict=model.predict(X_test)
test_predict_norm=scaler.inverse_transform(test_predict)
math.sqrt(mean_squared_error(ytest,test_predict_norm))
```
